```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def logisticMap(r, currentNumber):
    return r*currentNumber*(1-currentNumber)
```

```python
listOfX = []
listOfR = []

x0 = 0.5

t = np.arange(0.0, 4.0, 0.001) # Number of sample point distance
for n in range(len(t)):
    x = x0
    for i in range(1000): # Number of iterations to increase accuracy
        x = logisticMap(t[n], x)
    for i in range(1000): # Number of iterations added to final list
        x = logisticMap(t[n], x)
        listOfX.append(x)
        listOfR.append(t[n])
```

```python
plt.plot(listOfR, listOfX, marker = ',', alpha = 0.05, color = 'black') # Change scatter dot alpha
plt.savefig('bifurication_diagram.png', dpi = 300) # Change dpi to increase / decrease output image resolution
plt.show()

# To get  high quality output, set: SAMPLE_POINT_DIST to 0.0001
#                                   both ITERATION amounts to 1000
#                                   ALPHA to 0.05
#                                   DPI to 1200 
```
